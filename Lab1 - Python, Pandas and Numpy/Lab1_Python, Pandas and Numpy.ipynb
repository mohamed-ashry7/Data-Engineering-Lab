To loop over the list we can use the ```for``` loop as follows:

````python
list_newest= ['a','b','c','d','e']

for i in range(len(list_newest)):
    list_newest[i]= list_newest[i] +" hello"
print (list_newest)

# instead of using the loop, you can use a less verbose code by using list comprehension
````

Or we can use a less verbose code by using list comprehension.

````python
list_newest= ['a','b','c','d','e']
list_newest = [ element+ " hello" for element in list_newest ] #list comprehension
print (list_newest)
````

What if I would like to access index and value at the same time?

````python
for index, value in enumerate(list_newest):
    print(f"The pair: {index}: {value}")
````

You can also start at a different index.

````python
for a, b in enumerate(list_newest, 3): # start at index 101
    print(f"The pair: {a}: {b}")
````

Another way to get the pair would be using lists comprehension.

````python
tuple_list = [k for k in enumerate(list_newest)] #returns a tuple of index and value in a list
print(tuple_list)
````

To access two lists at the same time simultaneously, use the `zip` function

````python
numeric_list = [22,33,44,55,66,99]
for i_1,i_2 in zip(numeric_list, list_newest): #access both lists at the same time
    print(f"Mixed Pair-> {i_1}:{i_2}")
````

## **Reading Files in Python 3.X**

We will be working with this [file](https://github.com/ComputoCienciasUniandes/python-bootcamp/blob/master/tareas/hamlet.txt) for this example. 

To read a file in python do the following:

````python
hamletfile=open("hamlet.txt")
hamlettext=hamletfile.read() #a string holding the text in the file
hamletfile.close()
hamlettokens=hamlettext.split()#split with no arguments splits on whitespaces (returns a list of the words split on whitespaces)
print(len(hamlettokens))
````

As you might have observed above, there are around 32k words. Note that we close the file to avoid extra memory usage. Luckly there exists a keyword that automatically closes the file, and that is `with`

````python
with open("hamlet.txt") as hamletfile:
    hamlettext=hamletfile.read()#a string holding the text in the files
    hamlettokens=hamlettext.split()#returns a list
    print (len(hamlettokens))
````

````python
joinedHalmetText=",".join(hamlettokens) #join a list to make a string separated by commas
joinedHalmetText
````

There are two main data structures in Pandas:
1.    **Series:** a one-dimensional array of values with an index.
2.   **DataFrame:** a two-dimensional array of values with both a row and a column index.

### Querying

A useful thing is that you can dice\sort\etc the dataframe. Here we look for the sepal length that is > 4.5 .

````python
df_iris[ df_iris['SepalLengthCm']	 > 4.7  ]
````

This gives us Trues and Falses. Such a series is called a **mask**.

````python
df_iris.query("SepalLengthCm > 4.7")
````

In the next cell we create a mask and use it to "index" into the dataframe to get the rows we want.

````python
df_iris[df_iris.SepalLengthCm	 < 5]
````

If you want to combine these conditions, use the second form and put '()' brackets around each condition. The query uses a boolean AND. Each condition ceates a mask of trues and falses.

````python
df_iris[(df_iris.SepalLengthCm > 4.7) & (df_iris.SepalLengthCm < 5)]#there were none greater than 4.5!
````
