### **Line Graph**

### We will start with a line graph

### Params

### Funcs

set_xticklabels

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
x = np.arange(1,11,1)
y = [a**2 for a in x ]
y1 = [a**3 for a in x]
ax.plot(x,y,label = 'Quadratic',color= 'g', marker='^', linestyle='--'  )
ax.plot(x,y1,'ro--',label = 'Cubic'  )
ax.grid()
ax.legend()

fig
```

### Example
Plotting the stock prices of the tech companies. 

```python
google = pd.read_csv('./Datasets/GOOGL_data.csv')
fb     = pd.read_csv('./Datasets/FB_data.csv')
apple = pd.read_csv('./Datasets/AAPL_data.csv')
amazon = pd.read_csv('./Datasets/AMZN_data.csv')
microsoft = pd.read_csv('./Datasets/MSFT_data.csv')
google.head()
```

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
ax.plot(google.close,label='Google')
ax.plot(fb.close,label='Facebook')
ax.plot(amazon.close,label='Amazon')
ax.plot(microsoft.close,label='Microsoft')
ax.plot(apple.close,label='Apple')
ax.legend(loc='upper left')
ax.set_xticks(np.arange(0,1500,100))
ax.set_xticklabels(google.date.values[::100])
ax.grid()
```

### **Bar Graph**

```python
movie_scores = pd.read_csv('./Datasets/movie_scores.csv')
movie_scores
```

```python
movies_fig = plt.figure(figsize=(10, 5), dpi=300)
ax = movies_fig.add_subplot(1,1,1)
x = np.arange(1,len(movie_scores.Tomatometer)+1)
width = 0.2
ax.bar(height= movie_scores.Tomatometer , width=width,  x = x-0.3/2)
ax.bar(height =movie_scores.AudienceScore,width=width, x = x+0.3/2)
ax.set_xticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_xticks(x)
```

### For horizontal bar just flip (x,y), (Width, Height) and so on. 

```python
movies_fig = plt.figure(figsize=(10, 5), dpi=300)
ax = movies_fig.add_subplot(1,1,1)
y = np.arange(1,len(movie_scores.Tomatometer)+1)
width = 0.2
ax.barh(width= movie_scores.Tomatometer , height=width,  y = y-0.3/2)
ax.barh(width =movie_scores.AudienceScore,height=width, y = y+0.3/2)
ax.set_yticklabels(movie_scores.MovieTitle.values,fontsize=7)
ax.set_yticks(y)
```

### **Histogram**

The difference between Histogram and barplot is that barplot comparing many columns/attributes together.  
On the other side, Histogram shows the distribution of values for a specific column.

```python
hist_fig, ax = plt.subplots(1,1)
ax.hist(np.random.randn(100),bins=50)
ax.hist(np.random.randn(100),bins=50)
```

```python
df_titanic = pd.read_csv("Datasets/titanic.csv",index_col=0)
df_titanic.head()
```

```python
hist_fig, ax = plt.subplots(1,1)
max_age = max(df_titanic.Age.values)
min_age = min(df_titanic.Age.values)
age_bins= int((max_age-min_age)//10)
ax.hist(df_titanic.Age.values,bins=age_bins,color='k')
ax.grid()
```

#### ***2d histogram***

```python
plt.hist2d(np.random.randn(50),np.random.randn(50))
plt.colorbar()
```

### ***Boxplot***

Boxplot carries information about max, min, mean, first quartile, third quartile and the outliers. 

```python
box_fig =plt.figure()
ax = box_fig.add_subplot(1,1,1)
ax.boxplot([np.random.randn(20),np.random.randn(20),np.random.randn(20)],labels=['A','B','C'])
```

### ***Scatter Plot***

```python
box_fig =plt.figure()
ax = box_fig.add_subplot(1,1,1)
ax.scatter(np.random.randn(20),np.random.randn(20),label= 'F1')
ax.scatter(np.random.randn(20),np.random.randn(20),label= 'F2')
ax.scatter(np.random.randn(20),np.random.randn(20),label= 'F3')
ax.legend()
```

```python
data = pd.read_csv('Datasets/anage_data.csv')
data = data[np.isfinite(data['Maximum longevity (yrs)']) & np.isfinite(data['Body mass (g)'])]
data.head()
```

```python
classes = data.Class.unique()
fig,ax = plt.subplots(1,1,figsize=(10,10))
ax.set_xscale('log')
ax.set_yscale('log')
for class_ in classes:
    ax.scatter(data[data.Class==class_]['Maximum longevity (yrs)'],data[data.Class==class_]['Body mass (g)'],label=class_ )
    
ax.legend()
```

### ***Bubble Plot***

Bubble plot is just a scatter plot with the ability to add another 2 dimensions which can be viewed as the scale s and the color c

```python
fig = plt.figure(figsize=(20,10))
ax = fig.add_subplot(1,1,1)
sp = ax.scatter(np.random.randn(20),np.random.randn(20),s=np.abs(np.random.randn(20))*100,alpha=0.5,c= np.abs([np.random.randn(20)]))
fig.colorbar(sp)
```

### ***Barplot*** 

```python
#let's create a bar chart based on the mean/median/std/var/cov (covariance) that each gender pays
# the default estimator is the mean not the median 
df_tips =sns.load_dataset("tips")
bp = sns.barplot(x= 'sex', y='total_bill',data=df_tips,estimator=np.median)
bp.set(xlabel="Gender",ylabel="Bill")#renaming the x and y axes
# black lines drawn on the bars represent the 95% confidence interval (this can be con
# figured through optional arguments).
df_tips.head()
```

### **Distribution Plots**

Distribution plots, display the distribution for one variable.

The below dataset describes the driver's status, when the car crashed in a certain state in the United States. Our role is trying to identify several factors:

1. How likely it is that the driver was not distracted when the car crashed?
2. What is the correlation between the driver not being distracted and speeding when the car crashed?
3. What is the relation between the car crashing and every two factors in the dataset?

```python
df_car_crashes = sns.load_dataset("car_crashes")#seaborn has built in functions to retrieve datasets
df_car_crashes.head()
```

```python
sns.distplot(df_car_crashes['not_distracted'],bins=20,kde=True)
#plots the kernel density estimate 
```

### **Join Plots**

Join plots show the correlation between two variables. In the below chart, they show the correlation of the accident taking place such that the person was both speeding and not distracted.

```python
sns.jointplot(df_car_crashes['not_distracted'],df_car_crashes['speeding'],kind='reg')
```

### **Pair Plots** 

The pair plots show the relationship between every variable and every other variable.

```python
sns.pairplot(df_car_crashes)
```
