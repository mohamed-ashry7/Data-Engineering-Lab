### **Ex1**

For the titanic data, plot a scattered plot with the *Age* on the x-axis and the *Survival* count on the y-axis, such that the size of the circle is equivalent to the mean *Fare* of the ticket for the passengers of a certain age. 

```python
df_titanic = pd.read_csv("Datasets/titanic.csv",index_col=0)
df_titanic.head()
```

```python
plt.figure(figsize=(15,10))

ages_group = df_titanic.groupby('Age')['Survived'].sum()
fare_group = df_titanic.groupby('Age')['Fare'].mean()

ages = ages_group.index
survivors = ages_group.values 
plt.scatter(ages,survivors,s =  fare_group.values*10, c= np.random.randn(len(fare_group.values)))
```

### Ex4

*Using the FIFA dataset, create a bar chart which visualizes the average overall rating of the players based on their nationalities. Make sure that the x tick labels are visible. Give the plot a title, labels on both axes, and display only the top 10 nationality overall ratings.*


```python
fifa_df= pd.read_csv('./Datasets/fifa.csv')
```
